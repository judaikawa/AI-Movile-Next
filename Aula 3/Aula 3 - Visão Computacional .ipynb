Aqui podemos armazenar valores intermedirios entre as cores preto (valor 0) e branco (valor 255). Os valores intermedirios refere-se  intensidade de luz

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import urllib
from urllib.request import urlopen
% matplotlib inline
```

```python
def load_image(url, isBitLy=True):
    # Incluindo o inicio da url do bit.ly
    if isBitLy:
        url = "https://bit.ly/" + url
        
    # Busanco a imagem
    req = urllib.request.urlopen(url)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv2.imdecode(arr, -1) # Load it as it is
    return img
```

```python
# Lendo e exibindo uma imagem
charmander = load_image("2Ez1jn2")

# Exibindo a imagem
img = cv2.cvtColor(charmander, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
